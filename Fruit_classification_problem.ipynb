# Convolutional Neural Network (CNN)
A Convolutional Neural Network also known as (CNN), is an artificial neural network that has so far been most popularly used for analysing images. Although image analysis has been the most widespread use of CNNs they can be used for other data analysis or classification problems as well. Conceptually we can think of  CNNs as neural networks that have some specialisation for being able to pick out or detect patterns and make sense of them. This pattern detection is what makes CNNs ideal for image analysis.  CNNs are good when you need to preserve spatial information between features. Because another approach to image classification will just be to flatten the images and use a normal NN, but in this case, we will lose the spatial connection between adjacent pixels in the image.
![alt text](https://miro.medium.com/max/1920/0*4V0cYpmKNE5SPdVz.jpg)


Computers are unable to recognize or look at images the way we humans would. So, we would have to find a way to convert these images into numbers. There are several ways to do this when it comes to Image Processing; we discuss the one commonly used in Machine Learning.  An image of height=500, width=500 and depth=3 (where the scaling of brightness intensity is  in the range 0-255) is decomposed into various components which are the colours;  red, green, and blue, then these are placed on a gridded array of shape 500x500x3  ( the cells in the grid are known as pixels).


**Convolutional Neural Network Layers**


What differentiates it from a standard multi-layered neural network, is that a CNN has hidden layers called Convolutional layers and these layers are precisely what makes detect objects.
So what Convolutional layer do? Just like any other layer a Convolutional layer receives input images and then transforms inputs in some way outputs to them to the next layer. This transformation is a Convolutional operation. 

In each Convolutional layer, we need to specify the number of filters the layers should have. These filters are just what detects the patterns. To understand how these filters can detect features, think about what is going on a single given image, it has multiple shapes, textures, multiple edges, objects. One type of "pattern" that a filter could detect could be circles and some filter could detect edges and another could detect squares. Now, these simple and geometric filters are what we see at the start of our network, the deeper our network goes the more sophisticated these filters become. So in later layers rather than the simple geometric shapes as our filters may be able to detect specific objects like eyes, ears, hair, fur, feathers, scales, and beaks even. 

Operationally,  refer to the figure below showing  image to human perception  (left-1st panel **-from  left to right direction**) and gridded decomposed image (right -2st panel  **-from  left to right direction**), start at top left position, then filter moves across the image from left to right ( by steps size of 1 column) and top to bottom (by steps size of  1 row), each mini matrix  filter hovers above in the image, has its element multiplied by the corresponding counterpart from filter, then these are summed up to give single number which is placed into the filter map. The number of pixels we shift at each step (1 or more) is called the stride.
![alt text](https://static.packt-cdn.com/products/9781789138139/graphics/57b07b69-9550-4cda-a798-0578c8e30c74.png)
To understand what is actually happening here with these convolutional layers, we could conceptually think of these filters as torchlights, with different traits moving across the image. So each torchlight would be interested in finding a single specific thing. Probably what many overlook is that we convolve the image not only to detect features but to also reduce (downsample) that number of data points we are dealing with, because if we were to look at all the individual points(pixels) on an image that would be computationally expensive.

**Pooling**

Convolutional blocks are often times followed by a pooling block to reduce the activation spatial dimensions. Pooling helps in reducing memory consumption in deeper layers. It is also an important step to convert the spatial information into features. There are different pooling strategies. The most common are max-polling and average pooling. In all cases, pooling reduces an input “block” (receptive field) into a 1x1 output block, while keeping the depth unchanged. The reduction is done by selecting the max input activation (max-pooling) or by taking an average (average-pooling). Similar to convolution, a pooling block maps a receptive field to a single “pixel” in the output.

![alt text](https://miro.medium.com/max/570/1*S86gKd43MIYquHIeR9m8JQ.png)

Similar to convolution, a pooling block maps a receptive field to a single “pixel” in the output. For this reason, we can define a polling spatial size (2x2, 3x3, etc.) and stride. Usually, the stride is chosen to have non-overlapping receptive fields to achieve a reduction in spatial size.

To check how our model will perform on unseen data (test data), we create a validation set. We train our model on the training set and validate it using the validation set (standard machine learning practice).


```
def split_train_test(files_list, split=0.7,split2=0.2):

  """ 
   shuffle  & split the files names in to training (70%) 
   validation (20%) and test list(10%)
  """
  sh_f=np.random.permutation(len(files_list)) 
  shuffle_list=[files_list[ip] for ip in sh_f]
  split_index = int(split*len(files_list))
  split_index2=int((split+split2) *len(files_list))
  train_list = shuffle_list[0:split_index]

  val_list = shuffle_list[split_index:split_index2]
  
  test_list = shuffle_list[split_index2:]

  return train_list,val_list,test_list
```

# Define the model’s architecture

To code this in, we’ll use Keras sequential model. However, since we have many layers in our model, I’ll introduce a new way to specify the sequence. We’ll go through the code line-by-line, so that you can follow what we are doing exactly. Then, we call an ‘empty’ sequential mode


We’ll add to this empty model one layer at a time. The first layer is a convolutional layer with filter size 2x2, taking inputs of shape height,width,depth. The padding is the ‘same’.

$\verb|model.add(Conv2D(20, (2, 2), padding="same",input_shape=inputShape))|$


What this is doing is to add this layer to our empty sequential model using the function model.add(). The first number 32 refers to the number nodes. The next pair of numbers (2,2) refer to the filter dimensions. Then, we specify the padding which is ‘same’. Notice that we did not specify stride. This is because stride=1 is a default setting, and unless we want to change this setting, we need not specify it.

The chosen activation after the first CNN layer  is a ‘relu’. In subsequent layers we do not specify dimensions going into next layer since Keras can infer the input size from the output size of the previous layer.

The next layer is a max pooling layer with pool size 2 x 2 and stride 2 (in both dimensions). The next four layers look really similar,

$\verb|model.add(Conv2D(50, (5, 5), padding="same"))|$

$\verb|model.add(Activation("relu"))|$

$\verb|model.add(MaxPooling2D(pool_size=(2, 2), strides=(2, 2)))|$

Lastly, we have to code in our fully connected layer,  however, at this point, our neurons are spatially arranged in a cube-like format rather than in just one row. To make this cube-like format of neurons into one row, we have to first flatten it. We do so by adding a Flatten layer:
$\verb|model.add(Flatten())|$

Now, we have a dense (FC) layer of 512 neurons with relu activation

And lastly, we have a dense (FC) layer with 10 neurons and softmax activation.
And we’re done with specifying our architecture! 


NOTE: Because we are facing a two-class classification problem, i.e. a binary classification problem, we could have instead   ended our network with a sigmoid activation, so that the output of our network will be a single scalar between 0 and 1, encoding the probability that the current image is class 1 (as opposed to class 0).

```

def classifier_model(width, height, n_channels, classes):
		"""
    deep learning model for classifying two classes
		input:
		width: image with
		height:image height
    n_channels: image n_channels(RGB)

		"""
		# initialize the model
		model = Sequential()
		inputShape = (height, width, n_channels)
		# first set of CONV => RELU => POOL layers
		model.add(Conv2D(20, (2, 2), padding="same",input_shape=inputShape))
		model.add(Activation("relu"))
		model.add(MaxPooling2D(pool_size=(2, 2), strides=(2, 2)))
		# second set of CONV => RELU => POOL layers
		model.add(Conv2D(50, (5, 5), padding="same"))
		model.add(Activation("relu"))
		model.add(MaxPooling2D(pool_size=(2, 2), strides=(2, 2)))
  	# first (and only) set of FC => RELU layers
		model.add(Flatten())
		model.add(Dense(100))
		model.add(Activation("relu"))

		# softmax classifier
		model.add(Dense(classes))
		model.add(Activation("softmax"))

		# return the constructed network architecture
		return model
```

```
# initialize the model
print("compiling model...")
model =classifier_model(width=51, height=51, n_channels=3, classes=2)
opt = Adam(lr=INIT_LR, decay=INIT_LR / EPOCHS)
model.compile(loss="binary_crossentropy", optimizer=opt,
	metrics=["accuracy"])
 
# train the network
print("training network...")
```

In our model training we are expoiting the data generator we created and we able to model train in parallel processing.

```
MODEL = model.fit_generator(generator=train_generator
                            , validation_data=valid_generator,
                            epochs=EPOCHS,use_multiprocessing=True,
                    workers=6)
```

The Figures below show how the accuracy evolves over training epochs. The model performs extremely well in terms of predicting the images when we look at the accuracy & loss metrics but we most note there's a jump at around the epoch 5~ this suggests that the model is unable to correctly label the given image. One could ask as to why the  sudden jump happens in both the training data and in the validation data, well this is easily attributed this to the level of augmentation we applied, in some rare cases the augmentation completely  deformed (deconstructs) augments the image of fruits( of leaves- not completely implausible to blur leaf and it may seem like a fruit) (which are usually circular in shape) to a point it appears almost like the images of of the non-fruit objects which are usually broad branches or leaves. At the reason it appears in the validation data is because the weighting parameters sudden diverge from those most images previously passed through the model.

 Even though the validation data set has not augment but the model with the diverging weights from the augmented image cannot reconcile the input image with model output hence the loss function in both training data and validation data increase, instead of decreasing, but once the other images that have been extremely distorted are passed through the model order is restore the loss function declines and accuracy increases.
